```python
# 1. Download the dataset: Dataset and import the necessary libraries

# Import the necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 2. Load the dataset.

df = pd.read_csv("titanic.csv")
df
```

```python
# Displaying the first five rows of the dataset

df.head()
```

```python
# 3. Perform Below Visualizations.

# ● Univariate Analysis
# ● Bi - Variate Analysis
# ● Multi - Variate Analysis

# Univariate Analysis
# Common techniques used in univariate analysis include histograms, box plots, bar charts, and summary statistics 
# such as mean, median, and mode.

plt.hist(df['age'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
```

```python
# Bi - Variate Analysis
# Bivariate analysis techniques include scatter plots, line plots, correlation analysis, and contingency tables. 
# These methods help to identify patterns, trends, and dependencies between the two variables.

plt.scatter(df['age'],df['fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
```

```python
# Multi - Variate Analysis
# Multivariate analysis involves analyzing the relationship between three or more variables simultaneously. 
# Multivariate analysis techniques include techniques such as multivariate regression, principal component analysis (PCA), 
# factor analysis, cluster analysis, and discriminant analysis. These methods provide insights into the interdependencies 
# and interactions among the variables in the dataset.

# For this analysis, let's consider the relationship between the passenger's age, fare, and their survival status. 
# We can use a scatter plot with color-coded points to represent different survival outcomes.

plt.scatter(df['age'],df['fare'],c = df['survived'].replace({0:'red',1:'green'}))
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare with Survival Status')
```

```python
# 4. Perform descriptive statistics on the dataset.

# Descriptive statistics for numerical variables
num_stats = df.describe()
num_stats
```

```python
# Descriptive statistics for categorical variables
cat_stats = df.describe(include ='O')
cat_stats
```

```python
# Frequency counts for categorical variables
cat_counts = df.select_dtypes(include='O').nunique()
cat_counts
```

```python
# 5. Handle the Missing values.

df.isnull().sum()
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['age']
```

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embarked']
```

```python
df['deck'] = df['deck'].fillna(df['deck'].mode()[0])
df['deck']
```

```python
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
df['embark_town']
```

```python
# We can see that the missing values are handled
df.isnull().sum()
```

```python
# 6. Find the outliers and replace the outliers

sns.boxplot(df.age)
```

```python
upper_limit=df.age.quantile(0.99)
upper_limit
```

```python
lower_limit=df.age.quantile(0.01)
lower_limit
```

```python
df=df[(df.age<=upper_limit) & (df.age>=lower_limit)]
sns.boxplot(df.age)
```

```python
# 7. Check for Categorical columns and perform encoding.

# Identify categorical columns
categorical_cols = df.select_dtypes(include='object').columns
categorical_cols
```

```python
# Perform encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols)

# Print the encoded DataFrame
print(df_encoded)
```

```python
# 8. Split the data into dependent and independent variables.

# Y is dependent variable(target)
y = df_encoded['survived']
y.head()
```

```python
# X is independent variable or predictors

X = df_encoded.drop(columns = ['survived'],axis = 1)
X.head()
```

```python
# 9. Scale the independent variables

name = X.columns
name
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
X_scaled = scale.fit_transform(X)
X_scaled
```

```python
X = pd.DataFrame(X_scaled, columns = name)
X
```

```python
# 10. Split the data into training and testing

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```
